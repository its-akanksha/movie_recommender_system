Now formatting all the columns in the right format to obtain the final dataset 
that will only contain three columns, namely Movie_Id, Title, and Tags.
    We will merge all the other columns to form the appropriate tags column
    

```python
import ast #to help deal with the string dictionaries
```

```python
def convert(obj):
  genre = []
  for i in ast.literal_eval(obj):
    genre.append(i['name'])
  return genre
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
  names = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      names.append(i['name'])
      counter = counter + 1
    else:
      break
  return names
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
print(movies['cast'])
```

```python
def convert_dir(obj):
  director = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':    
      director.append(i['name'])
      break
  return director
```

```python
movies['crew'] = movies['crew'].apply(convert_dir)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_data = movies[['movie_id', 'title', 'tags']]
```

```python
new_data
```

# **Building the Model**

```python
## Taking the roots of all the words

from nltk.stem import PorterStemmer

def stem(text):
  ps = PorterStemmer()
  y = []

  for i in text:
    y.append(ps.stem(i))
  
  return (" ".join(y))

new_data['tags'] = new_data['tags'].apply(lambda x: stem(x))
```

```python
new_data['tags']
```

```python
new_data.head()
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x: x.lower())
```

```python
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_data['tags']).toarray()
```

```python
vectors
```

```python
len(vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):
  movie_index = new_data[new_data['title']== movie ].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]

  print("The 5 similar movies to your chosen movie are:")
  for i in movies_list:
    print(new_data.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman')
```

```python
recommend('John Carter')
```
